```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
img = cv2.imread("img2.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
```

```python
model = load_model("digit.model")
img.shape
```

```python
def bounding_box(image,box_size,stride):
    bbox = []
    for col in range(0,image.shape[1]-box_size[0],stride[0]):
         for row in range(0,image.shape[0]-box_size[1],stride[1]):
                bbox.append([row,row+box_size[1],col,col+box_size[0]])
    return np.array(bbox)
                
```

```python
def digit_detection(image,bboxes,model,size):
    new = np.zeros([image.shape[0],image.shape[1]],"uint8")
    for (x,y,x1,y1) in bboxes:
            img1 = image[x:y,x1:y1]
            img1 = cv2.resize(img1,(size[0],size[1]))
            img2 = img1.reshape(1,size[0],size[1],size[2])
            pr = np.max(model.predict(img2))
            if pr > .90:
                new[x:y,x1:y1] = 255*pr
    return new
```

```python
bboxes = bounding_box(img,(50,50),(10,10))
digit = digit_detection(img,bboxes,model,(28,28,1))
```

```python
plt.imshow(digit,cmap="gray")
```

```python
plt.imshow(img,cmap="gray")
```

```python
fig,axes = plt.subplots(15,15,figsize=(20,20))
axes = axes.ravel()

(cnts, _) = cv2.findContours(digit, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)
i = 0
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    roi=img[y:y+h,x:x+w].copy()
    ret,thresh = cv2.threshold(roi,140,255,cv2.THRESH_BINARY_INV)
    dilate = cv2.dilate(thresh,(3,3))
    img1 = cv2.resize(dilate,(28,28))
    axes[i].imshow(img1,cmap="gray")
    img2 = img1.reshape(1,28,28,1)
    predict = model.predict_classes(img2)
    predict_prob = model.predict(img2)
    axes[i].set_title("{},{}".format(predict,int(np.max(predict_prob)*100)),fontsize=13)
    axes[i].axis("off")
    i+=1
plt.subplots_adjust(hspace=.5)
```

```python
new = cv2.drawContours(img.copy(),cnts,-1,(0,0,255),4)
```
